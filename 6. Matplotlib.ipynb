```python
import matplotlib.pyplot as plt
import numpy as np

xpoints=np.array([0,1,2,3,4])
ypoints=np.array([6,10,14,12,9])

#plt.plot(xpoints,ypoints)
#plt.plot(xpoints,ypoints,linestyle="dotted",linewidth=4,color="red",markersize=12)
plt.plot(xpoints,ypoints,'b+--',markersize=6,linewidth=2,color="red",alpha=0.9)

plt.xlabel("X points")
plt.ylabel("Y points")
plt.title("Array Graph")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints=np.array([0,1,2,3,4])
ypoints=np.array([6,10,14,12,9])

#plt.plot(xpoints,ypoints)
#plt.plot(xpoints,ypoints,linestyle="dotted",linewidth=4,color="red",markersize=12)
plt.plot(xpoints,'b+--',markersize=6,linewidth=2,color="red",alpha=0.9,label="X")
plt.plot(ypoints,label="Y")
plt.xlabel("X points")
plt.ylabel("Y points")
plt.title("Array Graph",loc="left")
plt.grid()
plt.show()
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
# Show legend
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

#plt.legend(loc='best')
plt.legend(loc='upper right',fontsize="small",shadow=True)
```
